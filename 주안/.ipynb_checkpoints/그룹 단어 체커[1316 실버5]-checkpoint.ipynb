### 예제 출력 1
> 3

### 예제 입력 2
> 4<br>
aba<br>
abab<br>
abcabc<br>
a<br>

### 예제 출력 2
> 1

```python
"""
전체를 입력받은 뒤 테스트셋의 갯수 N과 N개의 문자열 리스트 A로 분할
결과변수 r초기화
"""
N,*A=open(0).read().split();r=0
"""
문자열 리스트 A의 값을 한 개씩 순회하여 진행
1번동작: 각 문자열중 첫 값을 꺼내온 뒤 세트에 담음
>> 문자열은 최소 한개의 값을 가짐, 세트는 중복값을 허용하지 않는 자료형
2번동작: 문자열의 각 문자를 순회하며 그룹체커가 있는지 없는지 체킹함
> 그룹체커
>> 3번동작: 현재 보고있는 값과 동일할경우 다음값 확인
>> 4번동작: 현재 보고있는 값을 바꾸고 기존 세트에 추가해줌
> 그룹체커 아님
>> 5번동작: 즉시 순회를 멈추고 결과값 -1 == 결과값을 안 더해주는 것과 동일한 동작
"""
for i in A:
    x=i[0];t={x} # 1번동작
    for j in i[1:]: # 2번동작
        if x==j:continue # 3번동작
        elif j in t:r-=1;break # 5번동작
        x=j;t|={j} # 4번동작
    r+=1
print(r)
```
