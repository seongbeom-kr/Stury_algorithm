### 예제 출력 1
> 0

### 예제 입력 2
> 7 7<br>
8<br>
10

### 예제 출력 2
> 1

```python
"""
입력값을 int형으로 변환한 뒤 각 a,b,c,d에 할당
기울기를 구하기 위한변수 d 선언
"""
a,b,c,n=map(int,open(0).read().split());d=n+1
g=(a*n+b-c*n)-(a*d+b-c*d) # 기울기
"""
기울기가 0보다 크거나 같고(음수가 아니며)
최초 시작 값이 O(n)의 정의를 만족하는 경우 1
아닌경우 0을 출력
"""
print([0,1][(g>=0)&(a*n+b<=c*n)])
```
