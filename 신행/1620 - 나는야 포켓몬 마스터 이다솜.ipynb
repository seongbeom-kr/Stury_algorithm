# 1620 - 나는야 포켓몬 마스터 이다솜
## 유형 : 자료구조, 문자열 처리, 해시테이블




```python
#n은 도감에 등록된 포켓몬의 수, m은 맞춰야 하는 문제의 수\
n, m = map(int, input().split())

# 도감을 저장할 딕셔너리를 초기화
dogam = {}

# 사용자가 찾으려는 포켓몬 이름 또는 번호를 저장할 리스트를 초기화합니다.
finders = []

# 포켓몬의 수만큼 반복하여 도감에 포켓몬의 이름과 번호를 저장
for i in range(1, n + 1):
    a = input()  # 포켓몬의 이름을 입력받습니다.
    dogam[i] = a  # 번호를 키로 하여 이름을 저장
    dogam[a] = i  # 이름을 키로 하여 번호를 저장
    
# 문제의 수만큼 반복하여 찾으려는 포켓몬 이름 또는 번호를 리스트에 저장
for _ in range(m):
    finders.append(input())

# 찾으려는 포켓몬 리스트를 순회하며 해당하는 포켓몬의 번호 또는 이름을 출력
for finder in finders:
    if finder.isdigit():  # 입력받은 값이 숫자라면 번호로 간주
        print(dogam[int(finder)])  # 번호에 해당하는 포켓몬 이름을 출력
    else:  # 예외사항
        print(dogam[finder])  # 이름에 해당하는 포켓몬 번호를 출력
```
