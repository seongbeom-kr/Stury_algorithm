```python
#n    , m    \
n, m = map(int, input().split())

#    
dogam = {}

#         .
finders = []

#        
for i in range(1, n + 1):
    a = input()  #   .
    dogam[i] = a  #     
    dogam[a] = i  #     
    
#          
for _ in range(m):
    finders.append(input())

#          
for finder in finders:
    if finder.isdigit():  #     
        print(dogam[int(finder)])  #     
    else:  # 
        print(dogam[finder])  #     
```
