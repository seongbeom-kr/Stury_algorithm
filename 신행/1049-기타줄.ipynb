# 1049 : 기타줄
## 유형 : Greedy Algorithm



```python
# 정답 코드
n, m = map(int, input().split())
pack_prices = [] # 6개 패키지 가격 담는 리스트
single_prices = [] # 1개 단일 가격 담는 리스트

# 브랜드별로 팩 가격과 단품 가격을 입력받아 각 리스트에 저장
for _ in range(m):
    pack, single = map(int, input().split())
    pack_prices.append(pack)
    single_prices.append(single)

# 가장 저렴한 팩 가격과 단품 가격을 찾음
min_pack_price = min(pack_prices)
min_single_price = min(single_prices)

# 최소 비용을 계산
# 팩으로만 구매하는 경우, 단품으로만 구매하는 경우, 혼합해서 구매하는 경우 중 최소값을 선택
price = min(
    (n // 6 + 1) * min_pack_price,  # 모두 팩으로 구매하는 경우 (부족한 부분도 팩으로 구매)
    (n // 6) * min_pack_price + (n % 6) * min_single_price,  # 팩과 단품을 혼합해서 구매하는 경우
    n * min_single_price  # 모두 단품으로 구매하는 경우
)

# 최소 비용 출력
print(price)
```

```python
# 답은 맞게 나오는데 틀린 코드
n, m = map(int, input().split())
packs = [] # 6개 세트 가격 집합
ones = [] # 1개 가격 집합
price = 0
# 브랜드 개수만큼 리스트에 담기
for _ in range(m):
    a, b = map(int, input().split())
    packs.append(a)
    ones.append(b)

min_pack_price = min(packs)
min_one_price = min(ones)


while n>=6:
    # 6개 이상일 때
    price += min_pack_price
    n-=6
# 6개 미만일 때
# 적어도 n개이기 때문에, 더 사도 상관없음
if n>0:
    price += min(min_pack_price, min_one_price * n)
    
print(price)
```
