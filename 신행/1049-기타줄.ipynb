```python
#  
n, m = map(int, input().split())
pack_prices = [] # 6    
single_prices = [] # 1    

#         
for _ in range(m):
    pack, single = map(int, input().split())
    pack_prices.append(pack)
    single_prices.append(single)

#       
min_pack_price = min(pack_prices)
min_single_price = min(single_prices)

#   
#   ,   ,      
price = min(
    (n // 6 + 1) * min_pack_price,  #     (   )
    (n // 6) * min_pack_price + (n % 6) * min_single_price,  #     
    n * min_single_price  #    
)

#   
print(price)
```

```python
#     
n, m = map(int, input().split())
packs = [] # 6   
ones = [] # 1  
price = 0
#    
for _ in range(m):
    a, b = map(int, input().split())
    packs.append(a)
    ones.append(b)

min_pack_price = min(packs)
min_one_price = min(ones)


while n>=6:
    # 6  
    price += min_pack_price
    n-=6
# 6  
#  n ,   
if n>0:
    price += min(min_pack_price, min_one_price * n)
    
print(price)
```
